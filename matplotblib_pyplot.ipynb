## plot()의 주요 param

- **color** 컬러
- **alpha** 투명도
- **linewidth** 선 굵기
- **linestyle** 선 스타일
- **marker** 마커 타입
- **markersize** 마커 크기
- **markerfacecolor** 마커 내부 컬러
- **markeredgecolor** 마커 테두리선 컬러
- **markeredgewidth** 마커 테두리선 굵기



---
**[라인스타일]**
- **linestyle** 설명
- **none** 라인 없음
- **:** 점선
- **--** 단속 선
- **-.** 선과 점
- **-**연속선
- **steps** 스텝



## Bar

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

x = np.array(['Q1', 'Q2', 'Q3','Q4'])
y = np.array([ 234.0, 254.7, 144.6, 317.6])
plt.bar(x,y,color = 'purple')
plt.show()
```

```python
sns.set_theme(style="darkgrid")
df = pd.DataFrame({'Type of Food': ['French Fries', 'Potato Chips', 'Bacon', 'Pizza', 'Chili Dog'], 
      'Number of Calories': [607, 542, 533, 296, 260]})
p = sns.barplot(x = 'Type of Food', y= 'Number of Calories', data = df)
p.set_title("Calories per 100g for different Foods")
```

```python
sns.set_theme(style="white")
fig, ax = plt.subplots(figsize = (8, 6)) 
plt.bar(df['Type of Food'], df['Number of Calories'], color = ['lightgray', 'lightgray', 'r', 'lightgray', 'lightgray'])

for i, v in enumerate(df['Number of Calories']):
    ax.text(i - .1, v - 50, str(v), color='white', fontweight='bold')
ax.yaxis.set_visible(False)
plt.show()
```

## Histogram

```python
x = np.random.randn(10000)
plt.hist(x,bins=20,color='green', density=True)
plt.show()
```

## Box plot

```python
x = np.random.randn(10000)*10+3
y = np.random.randn(10000)*10+5
z = np.random.randn(10000)*10+1
plt.boxplot([x, y, z], 0)
plt.show()
```

## Line plot

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='red',linestyle='-.',linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
```

## Pie

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Group1', 'Group2', 'Group3', 'Group4']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

```python
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

## Scatter

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.scatter(x,y,c='red',marker='o',alpha=0.5)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```

```python
fig=plt.figure()
axes = fig.add_axes([0,0,1,1]) # Left, Bottom, Width, Height
axes.plot(x,y,color='red',linestyle='-')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('LINE PLOT')
plt.show()
```

```python
#figsize인자를 통한 비율 조정
fig=plt.figure(figsize=(10,2)) # 가로, 세로 크기.
axes = fig.add_axes([0,0,1,1]) # Left, Bottom, Width, Height
axes.plot(x,y,color='red',linestyle='-')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('LINE PLOT')
plt.show()
```

```python
#하나의 패널에 다중 시각화 (범례 사용):
y = np.sin(x)
z = np.cos(x)

fig=plt.figure(figsize=(8,5), dpi=100) # 가로, 세로 크기. DPI 세팅.
axes = fig.add_axes([0,0,1,1]) # Left, Bottom, Width, Height
axes.plot(x,y,color='red',linestyle=':', label='Sin') # 범례를 위한 label 인자.
axes.plot(x,z,color='blue',linestyle='-.', label='Cos') # 범례를 위한 label 인자.
axes.legend(loc=0) # Top-Right에 범례 출력.
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('LINE PLOT')
plt.show()
```

```python
#여러개의 패널이 있는 다중 시각화 1:
fig=plt.figure(figsize=(12,4)) 
axes1 = fig.add_axes([0,0,0.3,1]) # Left, Bottom, Width, Height
axes2 = fig.add_axes([0.5,0,0.3,1]) # Left, Bottom, Width, Height
axes1.plot(x,y,color='red',linestyle=':')
axes2.plot(x,y,color='blue',linestyle='-.')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('SINE')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('COSINE')
plt.show()
```

```python
#여러개의 패널이 있는 다중 시각화 2
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(5,5))
# (0,0)
axes[0,0].plot(x,y,color='green',linestyle=':')
axes[0,0].set_xlabel('X')
axes[0,0].set_ylabel('Y')
axes[0,0].set_title('sine')

# (0,1)
axes[0,1].plot(x,y,color='red',linestyle='--')
axes[0,1].set_xlabel('X')
axes[0,1].set_ylabel('Y')
axes[0,1].set_title('sine')

# (1,0)
axes[1,0].plot(x,y,color='orange',linestyle='-.')
axes[1,0].set_xlabel('X')
axes[1,0].set_ylabel('Y')
axes[1,0].set_title('sine')

# (1,1)
axes[1,1].plot(x,y,color='blue',linestyle='-')
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')
axes[1,1].set_title('sine')

plt.tight_layout() # 서로 겹치는 현상을 피함.
plt.show()
```
